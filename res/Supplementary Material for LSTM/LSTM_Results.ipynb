##  SUPPLEMENTARY MATERIAL FOR PAPER "An Empirical Study on Soft Target and Label Smoothing in Text Classification with Crowd Annotations" - LSTM RESULTS

Research on label smoothing has shown that using label distribution as the training target can enhance both model quality and calibration. In the context of crowdsourcing, the empirical distribution over crowd labels, namely the soft target, offers a potentially better distribution than that obtained by label smoothing. This paper proposes and analyzes  the effect of applying soft target and label fusion methods on model calibration error. We show that soft targets improve on model calibration with respect to label smoothing across several NLP datasets and problems with different "noise" levels in the crowdsourced data collection.

In the original paper, we present our results on evaluating the effect of both soft targets and label smoothing in training the multi-layer perception and DistilBERT (a deep transformer model). Here, we further present our results on training the bidirectional long short-term memory (BiLSTM).

We trained BiLSTM with the cross-entropy loss using i) hard targets (one-hot encoded labels), ii) label smoothing, iii) the proposed soft and iv) semi-hard (sHard) targets (please see the original paper for details on each method).
To study the impact of label fusion methods, we experimented with the following fusion algorithms: i) Majority Voting (MV), ii) Dawid-Skene (D&S), which models worker reliability, and iii) GLAD, which further takes into account task difficulty. For label smoothing, we considered alpha values of 0.05 and 0.1 for label smoothing.

For every configuration (BiLSTM, training target), we performed grid search for the following hyperparameters: learning rate, class weights, embedding size, dropout rate. 

The network were trained using ADAM optimizer iterated up to 500 epochs. We set the batch size of 128.

We used the same metrics presented in the original paper.
